This graph is a little bit confusing so we add color so that we can differenciate between species.

```r
ggplot(data=penguins)+geom_point(mapping=aes(x=flipper_length_mm,y=body_mass_g,color=species))+ facet_wrap(~species)
```

```r
library("tidyverse")
library("ggplot2")
install.packages("palmerpenguins")
library("palmerpenguins")
```

```r
library("ggplot2")
ggplot(data=penguins)+geom_point(mapping=aes(x=flipper_length_mm,y=body_mass_g,color=species))+ facet_grid(sex~species)+ labs(title="Palmer Pengui: Body Mass vs Flipper Length")
```

```r
ggplot(data=penguins)+geom_point(mapping=aes(x=flipper_length_mm,y=body_mass_g,color=species))+annotate("text", x=220,y=3500,label="The Gentoos are the largest")+
 labs(title="Palmer Penguins: Body Mass vs Flipper Length")
```
