Now lets find the number and the type medals that each country has won, along with which sport has the most deciplined Athletes.

```python
fig = plt.figure()
Medals.replace('\'','', regex=True, inplace=True) 
medals_country = Medals.loc[:,'Team/NOC']
medals_Gold = Medals.loc[:,'Gold']
medals_Silver = Medals.loc[:,'Silver']
medals_Bronze = Medals.loc[:,'Bronze']
ax = fig.add_axes([0,0,1,1])
fig.set_size_inches(18, 10)
ax.bar(medals_country,medals_Gold,color = '#FFD700')
ax.bar(medals_country,medals_Silver,color = '#C0C0C0')
ax.bar(medals_country,medals_Bronze,color =  '#CD7F32')
plt.xticks(rotation='vertical')
plt.legend(['Gold','Silver','Bronze'])
plt.show()
```

```python
fig = plt.figure()
teams.replace('\'','', regex=True, inplace=True) 
Discipline = teams.loc[:,'Discipline']
sns.countplot(Discipline)
plt.xticks(rotation='vertical')
plt.show()
```
